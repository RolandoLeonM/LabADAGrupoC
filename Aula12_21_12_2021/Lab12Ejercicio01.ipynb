# Ejercicio01: Group Anagrams

Dada un array de strings strs, agrupe los anagramas. Puede devolver la respuesta en cualquier orden.

Un anagrama es una palabra o frase formada reordenando las letras de una palabra o frase diferente, normalmente usando todas las letras originales exactamente una vez.

**Input:** strs = ["eat","tea","tan","ate","nat","bat"]

**Output:** [["bat"],["nat","tan"],["ate","eat","tea"]]

# Explicaci贸n de soluci贸n

Para resolver se crea un diccionario (d) para los caracteres separados de las palabras, luego usamos esto como una clave y agregamos las palabras ordenadas o clasificadas, posteriormente se muestra los valores del diccionario donde fueron agregados las palabras o strings.

# C贸digo

```python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]
    # Dado un conjunto de strings, la funci贸n devuelve una lista
    # de listas de anagramas
        d = {}
        for words in strs:
            key = tuple(sorted(words))
            d[key] = d.get(key,[]) + [words]
        return d.values()
```
