# Ejercicio 05: Espera de la entrevista

Estás sentado en una fila de sillas, esperando una entrevista.
Al elegir al siguiente entrevistado, el entrevistador siempre elegirá entre el lado izquierdo o el lado derecho de la fila de sillas.
Además, el entrevistador sabe exactamente cuánto tiempo le llevará cada entrevista y, siendo parado, siempre elegirá el bando con el menor tiempo de entrevista. (Puede asumir que todos los tiempos son distintos).
Se le indica con un tiempo de entrevista, lo que significa que el entrevistador siempre lo elegirá por encima de la persona del lado opuesto. ¿Cuánto tiempo pasará antes de que el entrevistador llegue a usted?

**Entrada: [4, -1, 5, 2, 3]**

**Salida: 9**

```python
from collections import deque

def waitTime(SC):
# Dada una lista (cola doblemente enlazada) de valores
# de tiempos de atención la función retornará el tiempo
# de espera para ser atendido.

    if not SC:
        return 0
    temp = 0
    S = deque(SC)
    tam = len(S)+1

    for i in range(1,tam):
        left = S[0]
        right = S[len(S)-1]
        if left == -1 or right == -1:
            break
        if left < right:
            temp += S.popleft()
        else:
            temp += S.pop()
    return temp
```

```python
s = [4, -1, 5, 2, 3]
print(waitTime(s))
```
