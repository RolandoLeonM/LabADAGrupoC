# Ejercicio 02: Agregar mínimo para que los paréntesis sean válidos

Dada una cadena S de '(' y ') ', calcule el número mínimo de '(' o ')' que debe agregarse para que la cadena de paréntesis resultante sea válida.
Formalmente, una cadena de paréntesis es válida si y solo si:

* Es la cadena vacía, o
* Puede escribirse como AB (A concatenado con B), donde A y B son cadenas válidas, o
* Puede escribirse como (A), donde A es una cadena válida.

```python
def minAddToMakeValid(S):
# Dado una cadena de parentesis la función retorna
# el número de parententesis no válidos.

    if not S:
        return 0
    
    temp = []

    for i in S:
        if i =='(':
            temp.append(i)
        else:
            if len(temp) > 0 and temp[len(temp)-1] == '(':
                temp.pop(len(temp)-1)
            else:
                temp.append(i)
    return len(temp)

```

```python
s1 = "())"
rpta1 = minAddToMakeValid(s1)
print(rpta1)
s2 = "(()))("
rpta2 = minAddToMakeValid(s2)
print(rpta2)
s3 = "((("
rpta3 = minAddToMakeValid(s3)
print(rpta3)
s4 = "()))(("
rpta4 = minAddToMakeValid(s4)
print(rpta4)
```
