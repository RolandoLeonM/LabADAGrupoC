# Ejercicio 06. Fusionar k listas ordenadas

Se le da una matriz de k listas de listas vinculadas, cada lista vinculada se ordena en orden ascendente. Combine todas las listas vinculadas en una lista vinculada ordenada y devuélvala.

**Entrada: listas = [[1,4,5], [1,3,4], [2,6]]**

**Resultado: [1,1,2,3,4,4,5,6]**

```python
from queue import PriorityQueue

def listSort(L):
# Dado una matriz de listas la función
# muestra una lista ordenada ascendentemente.

    # crea una cola de prioridad
    pq = PriorityQueue()
    new_list = []
    # convierte matriz a una lista
    for l in L:
        for item in l:
            new_list.append(item)
    # agrega a cola de prioridad
    for e in new_list:
        pq.put(e)
    # imprime cola de prioridad
    while not pq.empty():
        item = pq.get()
        print(item, end=" ")
```

```python
matrix = [[1,4,5],[1,3,4], [2,6]]
listSort(matrix)
```
