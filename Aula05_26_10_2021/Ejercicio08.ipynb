## Solución:

Hallando la complejidad de tiempo:

![](images/Q08_solver.png)
![](https://drive.google.com/uc?export=view&id=1Ted7BECs0mEEmDFj5X14HG8CMgfMy01_)

Sumando tenemos:

$1+(\frac{n}{2}+1)+\frac{n}{2}=\frac{2n}{2}+2=n+2$

Por lo tanto la complejidad es $O(n)$ es lineal

Si realizamos el cálculo por deducción o exhaustivo para $n=100$.

|$iteraciones$    |$i$       |
|-----------------|----------|
| $1\rightarrow$  | $100$    |
| $2\rightarrow$  | $50$     |
| $3\rightarrow$  | $25$     |
| $4\rightarrow$  | $12$     |
| $5\rightarrow$  | $6$      |
| $6\rightarrow$  | $3$      |
| $7\rightarrow$  | $1$      |

Ahora, hacemos para $n=20$

|$iteraciones$    |$i$       |
|-----------------|----------|
| $1\rightarrow$  | $20$     |
| $2\rightarrow$  | $10$     |
| $3\rightarrow$  | $5$      |
| $4\rightarrow$  | $2$      |
| $5\rightarrow$  | $1$      |

Se puede observar que las iteraciones varian  de acuerdo a los resultados de $\left \lfloor \frac{n}{2} \right \rfloor$



## Código implementando en Java:


```java
/*
  Imprime las mitades de un numero (piso)
*/
public void halfNumber(int n){
    // Dado un número se imprime sus mitades
    // hasta llegar a 1
    for (int i = n; i >= 1; i=i/2){
        System.out.println(i);
    }
}
```

```java
halfNumber(20)
```
