## Solución:

Hallando la complejidad de tiempo:

Realizamos el cálculo por deducción o exhaustivo.

|$i$              |$p$             |
|-----------------|----------------|
| $1\rightarrow$  |  $0+1$         |
| $2\rightarrow$  | $1+2$          |
| $3\rightarrow$  | $1+2+3$        |
| $\vdots$        | $\vdots$       |
| $k\rightarrow$  | $1+2+3+...+k$  |

Como se puede observar, el tiempo de ejecución de las sentencias es:

$1+2+3+\cdots +k=\frac{k(k+1)}{2}$

Como $p \leqslant n$, para que se detenga debera ser $p > n$.

Entonces:

$$k > n$$

Reemplazando $k$:

$$\frac{k(k+1)}{2} > n$$

Tenemos una aproximación mediante las propiedades asintoticas:
$$k^{2} \approx n$$

$$k \approx \sqrt{n}$$

Por lo tanto la complejidad es $O(\sqrt{n})$ es raiz cuadrada


## Código implementando en Java:


```java
/*
  Imprime la suma mayor aproximada de n numeros enteros
*/
public void plusNumbersHigher(int n){
    // Dado un número que es la suma imprime los numeros
    // enteros que cubren a la suma del numero dado
    int p = 0;
    for (int i = 1; p <= n; i++){
        p = p + i;
        System.out.print(i + " + ");
    }
    System.out.print(" = " + p);
}
```
