## Solución:

Hallando la complejidad de tiempo:

Realizamos el cálculo por deducción o exhaustivo para el primer for:

```java
for(j = 1; j < n; j = j * 2){
    statement;
}
```

|$iteraciones$      |$i$              |potencia  de  2 |
|-------------------|-----------------|----------------|
|$1$                | $1\rightarrow$  | $2^{0}$        |
|$2$                | $4\rightarrow$  | $2^{1}$        |
|$3$                | $8\rightarrow$  | $2^{2}$        |
|$4$                | $16\rightarrow$ | $2^{3}$        |
|$5$                | $32\rightarrow$ | $2^{4}$        |
|$6$                | $64\rightarrow$ | $2^{5}$        |
|                   | $\vdots$        | $\vdots$       |
|                   | $k\rightarrow$  | $2^{k}$        |

Como se puede observar, el tiempo de ejecución de las sentencias es:

$1+2+4+8+16+32+64+\cdots +k$

La sumatoria será:

$$\sum_{n=1}^{k}{2^{n-1}}= 2^{k}-1$$

Como $i < n$, para que se detenga debera ser $i \geqslant n$.

Entonces:

$$k \geqslant n$$

Reemplazando $k$:

$$2^{k}-1 \geqslant n$$

Tenemos una aproximación mediante las propiedades asintoticas:
$$2^{k} \approx n$$

Aplicando logaritmo en base 2:

$$\log_{2}2^{k} \approx \log_{2}n$$

$$k(\log_{2}2) \approx \log_{2}n$$

Entonces tenemos que:

$$k \approx \log_{2}n$$

La complejidad será $O(\log_{2}{n})$

.

---

Entonces el **análisis del algoritmo** será:


![](images/Q12_solver.png)
![](https://drive.google.com/uc?export=view&id=1FinfrveLI-Ax4vAjT_Ifqb4JL2Exxm1d)

Sumando tenemos: $2n+2+(\log_{2}{n})n=n\log_{2}{n}+2n+2$

Mediante las propiedades asintóticas tenemos: $n\log_{2}{n}$

Por lo tanto la complejidad es $O(n\log_{2}{n})$ que es cuasi-lineal

## Código implementando en Java:


````java
/*
  Imprime las potencias de dos repetidas veces
*/
public void powerOfTwoRepeats(int n){
    // Dado un número imrpime repetidamente la potencia de 2
    // potencia de 2 de los menores al número dado.
    int count = 0;
    for (int i = 0; i < n; i++){
       for (int j = 1; j < n; j = j * 2){
           System.out.println("2^"+count+" = "+j);
       }
       System.out.println();
    }
}
````
