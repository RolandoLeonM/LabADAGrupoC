Sumando tenemos: 2n + 2 + (2n + 2) * n = 2n^2 + 4n + 2

Entonces la complejidad de for será: O(n^2) que es cuadrática

## Código implementando en Java:


```java
/*
  Imprime la tabla de multiplicar del 0 hasta el 12
*/
public void multiplicationTable(int n){
    // Dado un número imprime la tablas de multiplicar
    // desde el cero hasta el número menor del dado 
    for (int i = 0; i < n; i++){
        for (int j = 0; j < n; j++){
            System.out.println(i+"*"+j+"\t = "+j*i);
        }
        System.out.println();
    }
}
```
