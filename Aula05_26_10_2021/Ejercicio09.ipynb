## Solución:

Hallando la complejidad de tiempo:

Realizamos el cálculo por deducción o exhaustivo.

|$i$              |$p$             |
|-----------------|----------------|
| $1\rightarrow$  | $1*1$          |
| $2\rightarrow$  | $2*2$          |
| $3\rightarrow$  | $3*3$          |
| $\vdots$        | $\vdots$       |
| $k\rightarrow$  | $k*k$          |

Como se puede observar, el tiempo de ejecución de las sentencias es:

$1+2+9+\cdots +k^{2}=\sum_{k=1}^{n}{k^{2}}$

Como $i < n$, para que se detenga debera ser $i \geqslant n$.

Entonces:

$$k \geqslant n$$

Reemplazando $k$:

$$k^{2} \geqslant n$$

Tenemos una aproximación mediante las propiedades asintoticas:
$$k^{2} \approx n$$

$$k \approx \sqrt{n}$$

Por lo tanto la complejidad es $O(\sqrt{n})$ es raiz cuadrada


## Código implementando en Java:


```java
/*
  Imprime los cuadrados de un número
*/
public void plusNumbersHigher(int n){
    // Dado un número se imprime los cuadrados
    // de los números menores al número dado
    for (int i = 0; i * i < n; i++){
        System.out.println(i*i);
    }
}
```
