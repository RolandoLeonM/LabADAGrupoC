Entonces la complejidad de for será: O(n) que es lineal

## Código implementando en Java:


```java
/*
  Imprime los n primeros numeros en forma decreciente
*/
public void reverseNumber(int n){
    // Dado un número nos imprime una lista desde
    // el número dado hasta 1, en forma decreciente.
    for (int i = n; i > 0; i--){
        System.out.println(i);
    }
}
```
