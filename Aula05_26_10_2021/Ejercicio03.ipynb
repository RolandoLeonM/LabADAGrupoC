Sumando tenemos: 1 + n/2 + 1 + n/2 = 2n/2 + 2 = n + 2

Entonces la complejidad de for será: O(n) que es lineal

## Código implementando en Java:


```java
/*
  Imprime los números pares en forma creciente
*/
public void pairNumber(int n){
    // Dado un número imprime una lista de números
    // pares menores al número dado.
    for (int i = 0; i < n; i=i+2){
        System.out.println(i);
    }
}
```
