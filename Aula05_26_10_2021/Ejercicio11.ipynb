## Solución:

Hallando la complejidad de tiempo:

Realizamos el cálculo por deducción o exhaustivo para el primer for:

```java
p = 0;
for(i = 1; i < n; i = i * 2){
    p++;
}
```

|$iteraciones = p$  |$i$              |potencia  de  2 |
|-------------------|-----------------|----------------|
|$1$                | $1\rightarrow$  | $2^{0}$        |
|$2$                | $4\rightarrow$  | $2^{1}$        |
|$3$                | $8\rightarrow$  | $2^{2}$        |
|$4$                | $16\rightarrow$ | $2^{3}$        |
|$5$                | $32\rightarrow$ | $2^{4}$        |
|$6$                | $64\rightarrow$ | $2^{5}$        |
|                   | $\vdots$        | $\vdots$       |
|                   | $k\rightarrow$  | $2^{k}$        |

Como se puede observar, el tiempo de ejecución de las sentencias es:

$1+2+4+8+16+32+64+\cdots +k$

La sumatoria será:

$$\sum_{n=1}^{k}{2^{n-1}}= 2^{k}-1$$

Como $i < n$, para que se detenga debera ser $i \geqslant n$.

Entonces:

$$k \geqslant n$$

Reemplazando $k$:

$$2^{k}-1 \geqslant n$$

Tenemos una aproximación mediante las propiedades asintoticas:
$$2^{k} \approx n$$

Aplicando logaritmo en base 2:

$$\log_{2}2^{k} \approx \log_{2}n$$

$$k(\log_{2}2) \approx \log_{2}n$$

Entonces tenemos que:

$$k \approx \log_{2}n$$

La complejidad será $O(\log_{2}{n})$

**Comprobando**:

Cuando $n=10$, entonces el $\log_{2}10=\left \lceil 3.3219 \right \rceil = 4$ iteraciones

Para $p$ será el número de iteraciones, el segundo for dado por:

```java
for(j = 1; j < p; j = j * 2){
    statement;
}
```

El análisis es el mismo, por lo cual podemos decir que:

Sumando tenemos: $1+\log_{2}{n}+\log_{2}{n}=1+2\log_{2}{n}$

Mediante las propiedades asintóticas tenemos: $\log_{2}{n}$

Por lo tanto la complejidad es $O(\log_{2}{n})$ que es logarítmica

## Código implementando en Java:


````java
/*
  Imprime las potencias de dos
*/
public void powerOfTwo(int n){
    // Dado un número, cuenta los numeros de
    // potencia 2 de los menores al número dado,
    // luego imprime los numeros de potencia 2
    // de los menores al contador
    int count = 0;
    int p = 0;
    for (int i = 1; i < n; i = i*2){
        p++;
    }
    for (int j = 1; j < p; j = j*2){
        System.out.println("2^"+count+" = "+j);
        count++;
    }
}
````
