## Solución:

Hallando la complejidad de tiempo:

![](images/Q10_solver.png)
![](https://drive.google.com/uc?export=view&id=1SqsBAZxNPt2mylqYBWTczNHOq0EexNAM)

Entonces sumando tenemos:

$$2n+2+2n+2=4n+4$$

Mediante las propiedades asintoticas:
$$=4n$$

Por lo tanto la complejidad es $O(n)$ es lineal


## Código implementando en Java:


```java
/*
  Imprime listas de números repetidos
*/
public void generateEqualLists(int n){
    // Dado un número se imprime dos listas
    // de números repetidos
    for (int i = 0; i < n; i++){
        System.out.print(i+1+" ");
    }
    System.out.println();
    for (int j = 0; j < n; j++){
        System.out.print(j+1+" ");
    }
}
```

```java
generateEqualLists(20)
```
