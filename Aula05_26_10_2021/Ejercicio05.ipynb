## Solución:

Hallando la complejidad de tiempo:

![](images/Q05_solver.png)
![](https://drive.google.com/uc?export=view&id=1ptctBE_WDwu1ygllsru8_kgOvu0HVTB-) 

Sumando tenemos: 2n + 2 + (2i + 2) * n

Para este caso existe una dependencia de i en el segundo for.

Entonces se tiene que hacer un cálculo por deducción o exhaustivo.
$$
\begin{matrix}
 &i\rightarrow  &0  &1  &2  &3  &...   &n \\ 
 &j\rightarrow  &0  &0  &0  &0  & &0 \\ 
 &  &\times  &1  &1  &1  & &1 \\ 
 &  &  &\times  &2  &2  & &2 \\ 
 &  &  &  &\times  &3  & &3 \\ 
 &  &  &  &  &\times  & &\vdots  \\ 
 &  &  &  &  &  & &n 
\end{matrix}
$$

Como se puede observar, el tiempo de ejecución de las sentencias es:

$1+2+3+\cdots +n=\frac{n(n+1)}{2}$

Por lo tanto la complejidad es $O(n^2)$ es cuadrática


## Código implementando en Java:


```java
/*
  Imprime el factorial de n numeros
*/
public void factorial(int n){
    // Dado un número imprime los factoriales
    // menores al número dado
    int aux = 1;
    for (int i = 0; i < n; i++){
        for (int j = 0; j < i; j++){
            if(j == 0){
                aux = 1;
            }else{
                aux=aux*(j+1);
            }
        }
        System.out.println(i + "!\t = " + aux);
    }
}
```
