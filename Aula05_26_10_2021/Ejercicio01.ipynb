Sumando tenemos: 1 + n + 1 + n = 2n + 2

Entonces la complejidad de for será: O(n) que es lineal

## Código implementando en Java:


```java
/*
  Imprime los n primeros numeros
*/
public void primerosNumeros(int n){
    // Dado un número nos imprime una lista
    // desde el 1 hasta el numero dado.
    for (int i = 0; i < n; i++){
        System.out.println(i+1);
    }
}
```

```java
primerosNumeros(5)
```
