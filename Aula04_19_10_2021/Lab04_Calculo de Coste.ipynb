# Ejercicio 01: Convertir O(n^2) a O(n)
Convertir el siguiente algoritmo de O(n^2)
a un algoritmo secuencial O(n)

```python
# 1. Convert from 0(n^2) to 0(n)

# Algortimo de O(n^2)

def greatestNumber(array):
# Dado un conjunto de valores la función
# nos retorna el mayor número
    for i in array:
        isIValTheGreatest = True
        for j in array:
            if j > i:
                isIValTheGreatest = False
        if isIValTheGreatest:
            return i
```

```python
A=[1,5,7,8,2,15,6]
print(greatestNumber(A))
```

```python
# Algoritmo convertido a O(n)

def mayorNumber(arreglo):
# Dado un conjunto de valores la función
# nos retorna el mayor número
    mayor = arreglo[0]
    # Recorrer y buscar
    for elemento in arreglo:
        if elemento > mayor:
            mayor = elemento
    return mayor
```

```python
A=[1,5,7,8,2,15,6]
print(mayorNumber(A))
```

```python
# 2. What is the time complexity of this code?

def every_other(array)
# Dado un conjunto de valores la función imprime
# la suma del valor de cada índice par con
# cada uno de los valores del array
	array.each_with_index do |number, index|
		if index.even?
			array.each do |other_number|
				puts number + other_number
			end
		end
	end
end

```

```python
A = [1,2,3]
every_other(A)
```

# Ejercicio 03: ¿Cuál es la complejidad del tiempo en términos de O()?
El siguiente código esta en JavaScript

```python
// 3. What is the time complexity in terms of 0()?

function twoSum(array){
// Dado un conjunto de valores la función nos retorna
// verdadero (true) si la suma de sus valores sean igual
// a 10 y sus indices sean diferentes, sino será falso
	for (let i = 0; i < array.length; i++) {
		for (let j = 0; j < array.length; j++) {
			if (i !== j && array[i] + array[j] === 10){
				return true;
			}
		}
	}
return false;
}
```

```python
var arreglo = [1,2,3,4,5,6]
console.log(twoSum(arreglo))
```

# Ejercicio 04: Encontrar la aguja en el pajar 
#¿Cuál es la complejidad del tiempo en términos de O()?
El siguiente código esta en Julia


```python
# 4. What is the time complexity in terms of 0()?
# "fgh" "abcdefghi"
function find_needle(needle, haystack)
# Dado dos conjuntos de valores, la función nos retorna
# verdadero si los valores de needle se encuentran dentro
# de los valores de haystakc, sino retornará falso
	needle_index = 1
	haystack_index = 1
	found_needle = true

	while haystack_index < length(haystack)
		if needle[needle_index] == haystack[haystack_index]
			found_needle = true
			while needle_index < length(needle)
				if needle[needle_index] != haystack[haystack_index + needle_index]
					found_needle = false
				break
			    end
			    needle_index += 1
			end
		return true
		end
		if found_needle
			needle_index = 1
	    end
	    haystack_index += 1
	end
	return false
end
```

```python
array1 = ["f", "g", "h"]
array2 = ["a", "b", "c", "d", "e", "f", "g", "h", "i"]
println(find_needle(array1, array2))
```
