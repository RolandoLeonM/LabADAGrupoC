¿Cuál es la complejidad temporal en terminos de $O( )$?

```python
# 3. What is the time complexity in terms of 0()?

def twoSum(array):
# Dado un conjunto de valores la función nos retorna
# verdadero (true) si la suma de sus valores sean igual
# a 10 y sus indices sean diferentes, sino será falso
	for i in range (0,len(array)):
		for j in range(0,len(array)):
			if (i !=j ) and (array[i] + array[j] ==10):
				return True
	return False
```

```python
arreglo = [1, 2, 3, 4, 5, 6]
twoSum(arreglo)
```
