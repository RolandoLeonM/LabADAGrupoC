```python
def pick_resume(resumes):
# Dado un conjunto de valores la funcion retorna
# el elemento que queda al disminuir el arreglo
# tanto por la derecha e izquierda.
# Nota: no funciona con cantidades de elementos como:
# [],2,9...12,37...53, y otras canidades de elementos
    eliminate = "top"
    while len(resumes) > 1:
        if eliminate == "top":
            resumes = resumes[len(resumes) // 2 : len(resumes) - 1]
            eliminate = "bottom"
        elif eliminate == "bottom":
            resumes = resumes[0 : len(resumes) // 2]
            eliminate = "top"
    return resumes[0]
```

```python
array = list(range(1, 9))
print(array)
print(pick_resume(array))
```
