¿Cuál es la complejidad temporal de este código?

```python
def every_other(array):
# Dado un conjunto de valores la función imprime
# la suma del valor de cada índice par con
# cada uno de los valores del array
    for i in range (0, len(array)):
        if i % 2 == 0:
            for e in array:
                print(array[i] + e)
```

```python
A = [1,2,3]
every_other(A)
```
