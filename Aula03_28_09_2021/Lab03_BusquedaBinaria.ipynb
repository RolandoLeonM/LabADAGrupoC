#**Búsqueda Binaria de una lista**


```python
def busquedaBinaria(sorted_arr, element):
	i = 0
	start = 0
	end = len(sorted_arr) - 1

	while i < len(sorted_arr):
		middle = (start + end) // 2
		if sorted_arr[middle] == element:
			return True
		elif sorted_arr[middle] < element:
			start = middle + 1
		else:
			end = middle - 1
		i += 1
	return False
```

```python
listaPrueba = [1, 2, 3, 5, 5, 5, 7, 9, 10, 14, 15, 18, 40, 50, 50,]
print(busquedaBinaria(listaPrueba, 18))
print(busquedaBinaria(listaPrueba, 13))
```

# **Búsqueda Binaria de varias listas y gráfico**

```python
def searchBinary(lista, elemento):
	# De un cinjunto de numeros nos retorna verdadero
    # si el número fue encontrado o caso contrario falso
    
    # variable indice del array para iteracion
	i = 0

	# variables para rastrear el área de búsqueda 
	# inicialización de los índices de inicio y fin
	inicio = 0
	fin = len(lista) - 1

	# iterando sobre la lista 
	while i < len(lista):
		# obteniendo el índice del elemento medio (piso)
		middle = (inicio + fin) // 2

		# comprobando el elemento intermedio con el elemento requerido 
		if lista[middle] == elemento:
			## devuelve True si el elemento está en la lista 
			return True
        # caso contrario verifica si el elemento es mayor al elemento medio
		elif lista[middle] < elemento:
			## mueve el índice de inicio del área de búsqueda a la derecha 
			inicio = middle + 1
		else:
			## mueve el índice final del área de búsqueda a la izquierda 
			fin = middle - 1
		i += 1
	return False
```

```python
from random import randint

def generateValues(size):
    # Dada una cantidad generamos numeros aleatoriamente entre 1 a 100
    numberRandom = [randint(1,100) for _ in range(size)]
    # Ordenamos los valores generados
    numeros = sorted(numberRandom)
    return numeros
```

```python
import time
import matplotlib.pyplot as plt

# gráfica de la complejidad de search Binary
# acumula el rango de elementos que se buscaron 
elementos = list()
# acumula el tiempo de busqueda por rango
tiempos = list()
# tamaño de los primeros valores
start = 1000
# tamaño de busqueda de 10 en 10
tamano = 10_000 + start

for num in range(start, tamano, start):
# for i in range(tamano):
    # generando valores
    lista = generateValues(num)
    # tomando el tiempo por cada cantidad agregada y ordenada
    inicio = time.clock()
    # se realiza la búsqueda dada la lista
    buscando = searchBinary(lista, 15)
    fin = time.clock()
    # imprime el tiempo por la ordenación que realiza
    print(num, "Elementos buscados por searchBinary en ", fin-inicio," ",buscando)
    # agregando rango de valores
    elementos.append(num)
    # agregando a las lista
    tiempos.append(fin-inicio)
    i += 10
tiempos.sort()
# imprimiendo gráfica   
plt.xlabel('List Length')
plt.ylabel('Time Complexity')
plt.plot(elementos, tiempos, label ='Binary Search')
# plt.plot(tiempos, elementos, label ='Binary Search')
plt.grid()
plt.legend()
plt.show()
```

### **Búsqueda Binaria de matriz 2D**

```python
def searchBinaryMatrix(matrix, target):
	#m, n = len(matrix), len(matrix) and len(matrix[0])
	m = len(matrix)
	n = len(matrix[0])
	r, c = 0, n-1
	while r < m and c >= 0:
		if target > matrix[r][c]:
			r += 1
		elif target < matrix[r][c]:
			c -= 1
		else: return True
	return False
```

```python
matrix = [[1,3,5],[7,9,11],[13,15,17]]
searchBinaryMatrix(matrix,11)
```
