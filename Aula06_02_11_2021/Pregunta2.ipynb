# Pregunta 2

Encuentre el primer valor mayor o igual que x

```python
def greaterEqual(values, target):
# Dado un conjunto de valores y un elemento de búsqueda,
# la función retornará la posición del elemento encontrado
# o uno mayor al elemento, caso contrario retornará -1
    a = values
    N = len(a)
    L = 0
    R = N - 1
    higherEqual = -1
    while L <= R:
        mid = L + (R - L) // 2
        if a[mid] == target:
            return mid
        if a[mid] < target:
            L = mid + 1
        else:
            higherEqual = mid
            R = mid - 1
    return higherEqual
```

```python
x = 4
array = [2, 3, 5, 6, 8, 10, 12]
print(greaterEqual(array,x))
```
