Compare el elemento mid con el primer o el último elemento

```python
def minorElement(values):
# Dado un conjunto de valores la función retorna la
# posición del elemento menor encontrado, caso
# contrario retornará -1
    a = values
    N = len(a)
    L = 0
    R = N - 1

    while L <= R:
        mid = L + (R - L) // 2

 #       if a[mid] == a[L] or a[mid] == a[R]:
        if a[mid] == a[R]:
            return mid
 
        if a[mid] > a[R]:
            L = mid + 1
        else:
            R = mid
 
    return -1
```

## Casos de prueba

```python
array1 = [2, 3, 6, 7, 9, 15, 19]
print("posicion en array1:", minorElement(array1))
```
