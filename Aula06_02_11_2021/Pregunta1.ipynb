 ¿Es x un cuadrado? No use la función sqrt. Tienen que usar búsqueda binaria

```python
def isSquare(value):
# Dado un valor la función retorna "yes" si es un
# cuadrado, caso contario "no"

    L = 0
    R = value - 1
    while L <= R:
        mid = int(L + (R - L) / 2)
        if mid*mid == value:
            return "yes"
        if mid*mid < value:
            L = mid + 1
        else:
            R = mid - 1
    return "no"
```

```python
x = 16
print(isSquare(x))
```
