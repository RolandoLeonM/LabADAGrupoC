#**Búsqueda Secuencial**
##Realizar el algoritmo de búsqueda secuencial donde se ingrese una lista de números y también el valor, deberá retornar verdadero si encuentra el valor y falso caso contrario

```python
import time
import matplotlib.pyplot as plt
#from pylab import plot, show
```

```python
def sequentialSearch(numbers, target):
# De un conjunto de números retorna verdadero
# si el valor se ha encontrado o sino falso
    n = len(numbers)
    for i in range(n):
        if numbers[i] == target:
            return True
    return False
```

```python
values = [4, 5, 12, 45, 78, 15, 36,28, 11, 69]
search = sequentialSearch(values, 15)
print(search)
```

```python
def generateValues(size):
# Genera lista de números de acuerdo al tamaño dado
    # lista creciente desde 1 hasta el tamaño dado
    #return list(range(1, size+1))
    # lista decreciente desde el tamaño dado hasta 1
    return list(range(size, 0, -1))
```

```python
val = generateValues(10_000)
# print(val)
timeIni = time.perf_counter()
search = sequentialSearch(val, 15)
timeFin = time.perf_counter()
print(f"Tiempo transcurrido {timeFin - timeIni:0.6f} segundos")
```

```python
# gráfica de la complejidad de insert sort
elementos = list()
tiempo = list()
tiempos = list()
numero = 0
datos = list()
listita = generateValues(20)
print(listita)

for i in listita:
    # pasando un valor de la lista
    numero = i
    # agregando cada valor a la lista
    datos.append(numero)
    # tomando el tiempo por cada valor agregado y ordenado
    inicio = time.clock()
    buscando = sequentialSearch(datos, 15)
    fin = time.clock()
    # imprime el tiempo por la ordenación que realiza
    print(len(datos), "Elemento buscado por sequentialSearch en ", fin-inicio," ",buscando)
    # agregando a las listas
    elementos.append(len(datos))
    tiempo.append(fin-inicio)
    tiempos.append(sum(tiempo))

#print(tiempos)
#plot(elementos, tiempos)
#show()
plt.xlabel('List Length')
plt.ylabel('Time Complexity')
plt.plot(elementos, tiempos, label ='Sequential Search')
plt.grid()
plt.legend()
plt.show()
```

##**Ordenamiento por Inserción**
# Realizar el algorirmo de ordenamiento por inserción de una lista de valores enteros aleatorios

```python
import random
import time
import matplotlib.pyplot as plt
```

```python
def insertion_Sort(A):
# De un conjunto de enteros desordenados, se comparan los valores y sólo se
# intercambian de posición para su ordenación mientras se van desplazando.
    j = 1
    n = len(A)
    for j in range(n):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i] > key:
            A[i + 1] = A[i]
            i -= 1
        A[i+1] = key
    #print(A) # también cambia el arreglo exterior
```

```python
insertion_Sort(values)
print(values)
```

```python
def generateValuesRandom(size):
# Genera lista de números aleatorios de acuerdo al tamaño dado,
# determinado con valor incial, valor final y salto o paso
    lista = [random.randrange(0, 10_000, 10) for _ in range(size)]
    return lista
```

```python
#print("Ingrese cuantos numeros aleatorios desea obtener")
#n=int(input())
aleatorios=generateValuesRandom(10)
print(aleatorios)
timeIni = time.perf_counter()
insertion_Sort(aleatorios)
timeFin = time.perf_counter()
print(aleatorios)
print(f"Tiempo transcurrido {timeFin - timeIni:0.6f} segundos")
```

```python
# gráfica de la complejidad de insert sort
elements = list()
times = list()
num = 0
a = list()
lista = generateValuesRandom(10)
for i in lista:
    # pasando un valor de la lista
    num = i
    # agregando cada valor a la lista
    a.append(num)
    # tomando el tiempo por cada valor agregado y ordenado
    start = time.clock()
    insertion_Sort(a)
    end = time.clock()
    # imprime el tiempo por la ordenación que realiza
    print(len(a), "Elements Sorted by insertSort in ", end-start)
    # agregando a las listas
    elements.append(len(a))
    times.append(end-start)
times.sort()
print()
plt.xlabel('List Length')
plt.ylabel('Time Complexity')
plt.plot(elements, times, label ='Insert Sort')
plt.grid()
plt.legend()
plt.show()
```
