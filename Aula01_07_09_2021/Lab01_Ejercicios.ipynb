# Ejercicio 1

```python
def sum_numbers(values):
    """
    Dado un conjunto de numeros
    la funciÃ³n retorna la suma de los mismos
    """
    total = 0
    for n in values:
        total = total + n
    return total
```

```python
numbers = [1,2,3,4]
val = sum_numbers(numbers)
print(val)
```

```python
numbers = list(range(100))
val = sum_numbers(numbers)
print(val)
```

# Ejercicio 2

```python
def inverse(values):
    """
    Dado un conjunto de valores la funcion retorna
    un array con los valores en orden inverso
    """
    list_inv = []
    length = len(values)
    for i in range(length -1,-1,-1):
        list_inv.append(values[i])
    return list_inv
```

```python
numbers = [1,3,5,7,9]
val = inverse(numbers)
print(val)
```

# Ejercicio 3

```python
def parImpar(values):
    """
    Dado un conjunto de valores la funcion retorna
    dos array, uno con pares y el otro con impares
    """
    par = []
    impar = []
    for n in values:
        if n % 2 == 0:
            par.append(n)
        else:
            impar.append(n)
    return (par, impar)
```

```python
numbers = [1,2,3,4,5,6,7]
val = parImpar(numbers)
pares = val[0]
impares = val[1]
print("Pares:",pares)
print("Impares:",impares)
```

# Ejercicio 4

```python
def ordenamientoBurbuja(values):
    """
    Dado un conjunto de valores la funcion ordena
    en forma ascendente, mediante metodo de burbuja
    """
    for n in range(len(values)-1,0,-1):
        for i in range(n):
            if values[i] > values[i+1]:
                temp = values[i]
                values[i] = values[i+1]
                values[i+1] = temp
            
    return values
```

```python
numbers = [9,5,8]
val = ordenamientoBurbuja(numbers)
print(val)
```

# Ejercicio 5

```python
def frameText(text):
    """
    Dado un texto, la funcion imprime el
    texto dentro de un marco de asteriscos
    """
    line = (len(text)+4) * '*'+'\n'
    space = '* ' + (len(text)+1) * ' '+'*\n'
    middle = '* '+ text +' *\n'
    print(line + space + middle + space + line)
```

```python
text = "Hello world!"
frameText(text)
```
