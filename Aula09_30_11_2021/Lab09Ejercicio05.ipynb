# Ejercicio 03: Maximal Square
Dada una matriz binaria m x n llena de ceros y unos, encuentre el cuadrado m치s grande que contenga solo unos y devuelva su 치rea.

```python
def maximalSquare(self, matrix: List[List[str]]) -> int:
    # Dado una matriz de ceros y unos, la funci칩n retorna
    # el 치rea maxima del grupo de 1 que forma un cuadrado
    if not matrix: return 0
    R = len(matrix) #filas
    C = len(matrix[0]) # columnas
    dp = [[0]*C for _ in range(R)]
    max_count = 0
    for i in range(R):
        for j in range(C):
            dp[i][j] = int(matrix[i][j])
            if dp[i][j]: max_count = 1
    for i in range(1,R):
        for j in range(1,C):
            if dp[i][j] and dp[i-1][j] and dp[i][j-1] and dp[i-1][j-1]:
                dp[i][j] = min(dp[i-1][j-1],dp[i-1][j],dp[i][j-1])+1
                max_count = dp[i][j] if dp[i][j]>max_count else max_count
        return max_count**2
```
